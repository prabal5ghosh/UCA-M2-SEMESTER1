```python
import torch
import torchvision.models as models

model = models.resnet50(pretrained=True)  # Example using ResNet
model.eval()  # Set the model to evaluation mode
```

```python
from torchvision import transforms
from PIL import Image

preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])


```

```python
img = Image.open("dog1.jpg")
img_tensor = preprocess(img).unsqueeze(0)  # Add batch dimension
```

```python
img
```

```python
output = model(img_tensor)
_, predicted_class = torch.max(output, 1)
```

```python
from torch.autograd import Function

def get_gradients_hook(layer):
    def hook_function(module, grad_in, grad_out):
        layer.gradients = grad_out[0]
    return hook_function

target_layer = model.layer4[-1]  # Example with the last layer of ResNet50
target_layer.register_backward_hook(get_gradients_hook(target_layer))
```

```python
loss = output[0, predicted_class]  # Target class's score
model.zero_grad()
loss.backward(retain_graph=True)
```
