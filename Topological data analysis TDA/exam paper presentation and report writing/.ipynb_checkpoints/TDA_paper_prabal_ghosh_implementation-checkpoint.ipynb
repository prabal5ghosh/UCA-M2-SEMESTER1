```python
import numpy as np
import scipy.optimize as opt
import random
```

```python
class Tripeptide:
    def __init__(self, angles=None):
        """
        Represents a single tripeptide with 12 degrees of freedom.
        """
        self.angles = angles if angles is not None else np.random.uniform(-180, 180, 12)

    def apply_rotation(self, perturbation):
        """
        Applies a small random perturbation to the angles.
        """
        self.angles += perturbation
        self.angles = np.clip(self.angles, -180, 180)

class ProteinLoop:
    def __init__(self, num_tripeptides):
        """
        Initializes a loop as a sequence of tripeptides.
        """
        self.tripeptides = [Tripeptide() for _ in range(num_tripeptides)]

    def get_configuration(self):
        """
        Returns the full loop configuration as a flattened array.
        """
        return np.concatenate([t.angles for t in self.tripeptides])
```

```python
def tripeptide_loop_closure(tripeptides):
    """
    Solves the loop closure constraint using numerical optimization.
    """
    # def closure_constraint(angles):
    #     """
    #     Computes the difference between start and end points.
    #     """
    #     return np.linalg.norm(angles[:6] - angles[-6:])  # Ensure loop closure

    def closure_constraint(angles, tolerance=0.5):
        """
        Computes the difference between start and end points.
        Allows small variations to improve success rate.
        """
        return np.linalg.norm(angles[:6] - angles[-6:]) - tolerance  # Allow small deviation


    initial_guess = np.concatenate([t.angles for t in tripeptides])
    # result = opt.minimize(closure_constraint, initial_guess, method="BFGS")
    result = opt.minimize(closure_constraint, initial_guess, method="L-BFGS-B")


    if result.success:
        optimized_angles = result.x
        for i, t in enumerate(tripeptides):
            t.angles = optimized_angles[i*12:(i+1)*12]
    return result.success
```

```python
def hit_and_run_sampling(loop, steps=1000, step_size=5.0):
    """
    Uses the Hit-and-Run sampling method to explore the loop conformational space.
    """
    current_state = loop.get_configuration()
    accepted_samples = []

    for _ in range(steps):
        # Generate a random direction
        direction = np.random.randn(len(current_state))
        direction /= np.linalg.norm(direction)  # Normalize

        # Generate a step along the direction
        perturbation = direction * step_size

        # Apply perturbation
        new_loop = ProteinLoop(len(loop.tripeptides))
        for i, t in enumerate(new_loop.tripeptides):
            t.angles = loop.tripeptides[i].angles + perturbation[i*12:(i+1)*12]

        # Check for loop closure
        if tripeptide_loop_closure(new_loop.tripeptides):
            accepted_samples.append(new_loop.get_configuration())

    return np.array(accepted_samples)

```

```python
import numpy as np

def metropolis_acceptance(old_energy, new_energy, temperature=0.1):
    """
    Accept new conformation based on energy minimization.
    """
    if new_energy < old_energy:
        return True  # Always accept if energy decreases
    else:
        prob = np.exp((old_energy - new_energy) / temperature)
        return np.random.rand() < prob  # Accept with some probability
```

```python
# Create a loop with 5 tripeptides (15 amino acids)
protein_loop = ProteinLoop(num_tripeptides=5)

# Perform sampling
samples = hit_and_run_sampling(protein_loop, steps=100, step_size=2.0)


print(f"Generated {len(samples)} valid loop conformations.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_rmsf(samples):
    """
    Computes the Root Mean Square Fluctuations (RMSF) of heavy atoms across sampled conformations.
    samples: (N, D) array, where N = number of conformations, D = total degrees of freedom
    """
    mean_conf = np.mean(samples, axis=0)  # Compute mean structure
    rmsf = np.sqrt(np.mean((samples - mean_conf) ** 2, axis=0))  # RMSF calculation
    return rmsf

# Compute RMSF for generated samples
rmsf_values = compute_rmsf(samples)

# Plot RMSF as a boxplot
plt.figure(figsize=(10, 5))
plt.boxplot(rmsf_values.reshape(-1, 12))  # Reshape per tripeptide
plt.xlabel("Tripeptide Index")
plt.ylabel("RMSF (Å)")
plt.title("RMSF of Protein Loop Across Sampled Conformations")
plt.show()
```

```python
from sklearn.decomposition import PCA

def plot_pca(samples):
    """
    Projects sampled conformations into 2D using PCA for visualization.
    """
    pca = PCA(n_components=2)  # Reduce to 2D
    reduced_data = pca.fit_transform(samples)

    plt.figure(figsize=(8, 6))
    plt.scatter(reduced_data[:, 0], reduced_data[:, 1], alpha=0.5, color='blue')
    plt.xlabel("PC1 (Max Variance)")
    plt.ylabel("PC2")
    plt.title("PCA Projection of Sampled Loop Conformations")
    plt.show()

plot_pca(samples)
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_conformations(samples, num_to_plot=10):
    """
    Plots a subset of loop conformations in 3D.
    """
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')

    for i in range(min(num_to_plot, len(samples))):
        loop_coords = samples[i].reshape(-1, 3)  # Assuming each set of 3 values represents x,y,z
        ax.plot(loop_coords[:, 0], loop_coords[:, 1], loop_coords[:, 2], alpha=0.5)

    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.set_title("Sampled Loop Conformations (3D)")
    plt.show()

plot_3d_conformations(samples)
```

```python

def extract_loop_from_pdb(pdb_file, chain_id, loop_residues):
    """
    Extracts atomic coordinates for a given loop from a PDB file.
    :param pdb_file: Path to PDB file
    :param chain_id: Chain identifier (e.g., 'A')
    :param loop_residues: List of residue indices defining the loop
    :return: List of (x, y, z) atomic coordinates
    """
    parser = Bio.PDB.PDBParser(QUIET=True)
    structure = parser.get_structure("protein", pdb_file)
    chain = structure[0][chain_id]

    loop_coords = []
    for residue_id in loop_residues:
        if residue_id in chain:
            ca_atom = chain[residue_id]['CA']  # Use alpha carbon (Cα)
            loop_coords.append(ca_atom.coord)

    return np.array(loop_coords)

# Example Usage
pdb_file = "4ge2.pdb"  # Replace with your PDB file
chain_id = "A"
loop_residues = list(range(50, 1000))  # Example loop region

loop_coords = extract_loop_from_pdb(pdb_file, chain_id, loop_residues)
print(f"Extracted {len(loop_coords)} loop residues.")
```

```python
def train_on_real_data(pdb_file, chain_id, loop_residues, steps=1000):
    """
    Loads loop from PDB, applies Hit-and-Run sampling, and returns sampled conformations.
    """
    loop_coords = extract_loop_from_pdb(pdb_file, chain_id, loop_residues)

    # Reshape loop_coords to be compatible with ProteinLoop
    num_tripeptides = len(loop_coords) // 12  # Assuming 12 coordinates per tripeptide
    loop_coords = loop_coords[:num_tripeptides * 12].reshape(num_tripeptides, 12)

    # Create a ProteinLoop instance and initialize its angles
    protein_loop = ProteinLoop(num_tripeptides)
    for i, tripeptide in enumerate(protein_loop.tripeptides):
        tripeptide.angles = loop_coords[i]

    # Perform sampling using the ProteinLoop object
    samples = hit_and_run_sampling(protein_loop, steps)
    return samples
```

```python
from Bio.PDB import PDBIO, Structure, Model, Chain, Residue, Atom

def save_sampled_loops_as_pdb(samples, output_pdb):
    """
    Saves sampled loops as a PDB file for visualization.
    """
    structure = Structure.Structure("sampled_loops")
    model = Model.Model(0)
    chain = Chain.Chain("A")

    for i, sample in enumerate(samples[:10]):  # Save first 10 samples
        res_id = i + 1
        residue = Residue.Residue((" ", res_id, " "), "GLY", " ")  # Use glycine as placeholder

        for j, coord in enumerate(sample.reshape(-1, 3)):
            atom = Atom.Atom(f"CA{j}", coord, 1.0, 1.0, " ", f"CA{j}", "C")
            residue.add(atom)

        chain.add(residue)

    model.add(chain)
    structure.add(model)

    io = PDBIO()
    io.set_structure(structure)
    io.save(output_pdb)

# Save to a PDB file
save_sampled_loops_as_pdb(samples, "sampled_loops.pdb")
```
